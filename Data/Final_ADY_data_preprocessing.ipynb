```python
df_encoded = pd.get_dummies(df, columns=['Location', 'Fuel_Type', 'Transmission', 'Owner_Type','Brand'], drop_first=True)
df_encoded = df_encoded.replace({False: 0, True: 1})
df_encoded.head()
```

```python
df_encoded.info()
```

```python
Correct_Mileage= []
for i in df_encoded.Mileage:
    if str(i).endswith('km/kg'):
        i = i[:-6]
        i = float(i)*1.40
        Correct_Mileage.append(float(i))
    elif str(i).endswith('kmpl'):
        i = i[:-6]
        Correct_Mileage.append(float(i))
```

```python
df_encoded['Mileage']=Correct_Mileage
```

```python
# prompt: convert enginge column from string to int by delete CC

Correct_Engine= []
for i in df_encoded.Engine:
  if str(i).endswith('CC'):
    i = i[:-3]
    Correct_Engine.append(int(i))
df_encoded['Engine']=Correct_Engine
df_encoded.head()
```

```python
# prompt: # prompt: convert Power column from string to float by delete bhp

Correct_Power = []
for i in df_encoded.Power:
  if str(i).endswith('bhp'):
    i = i[:-4]
    Correct_Power.append(float(i))
df_encoded['Power'] = Correct_Power
df_encoded.head()
```
